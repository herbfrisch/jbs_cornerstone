```python
# An indicator variable (or dummy variable) is a numerical variable used to label categories. 
# They are called 'dummies' because the numbers themselves don't have inherent meaning.
# So we can use categorical variables for regression analysis later. 

cdf.columns
```

```python
# get indicator variables and assign it to data frame "dummy_variable_1"
dummy_variable_1 = pd.get_dummies(cdf["WEATHER"])
dummy_variable_2 = pd.get_dummies(cdf["ROADCOND"])
dummy_variable_3 = pd.get_dummies(cdf["LIGHTCOND"])
dummy_variable_4 = pd.get_dummies(cdf["SPEEDING"])

# change column names for clarity
dummy_variable_1.rename(columns={'WEATHER':'Clear', 'WEATHER':'Raining'}, inplace=True)
dummy_variable_1.rename(columns={'WEATHER':'Overcast', 'WEATHER':'Unknown'}, inplace=True)
dummy_variable_1.rename(columns={'WEATHER':'Snowing', 'WEATHER':'Other'}, inplace=True)
dummy_variable_1.rename(columns={'WEATHER':'Fog/Smog/Smoke', 'WEATHER':'Sleet/Hail/Freezing Rain'}, inplace=True)
dummy_variable_1.rename(columns={'WEATHER':'Blowing Sand/Dirt','WEATHER':'Severe Crosswind'}, inplace=True)
dummy_variable_1.rename(columns={'WEATHER':'Partly Cloudy'}, inplace=True)

dummy_variable_2.rename(columns={'ROADCOND':'Dry', 'ROADCOND':'Wet'}, inplace=True)
dummy_variable_2.rename(columns={'ROADCOND':'Unknown', 'ROADCOND':'Ice'}, inplace=True)
dummy_variable_2.rename(columns={'ROADCOND':'Snow/Slush', 'ROADCOND':'Other'}, inplace=True)
dummy_variable_2.rename(columns={'ROADCOND':'Standing Water', 'ROADCOND':'Sand/Mud/Dirt'}, inplace=True)
dummy_variable_2.rename(columns={'ROADCOND':'Oil'}, inplace=True)

dummy_variable_3.rename(columns={'LIGHTCOND':'Daylight', 'LIGHTCOND':'Dark - Street Lights On'}, inplace=True)
dummy_variable_3.rename(columns={'LIGHTCOND':'Unknown', 'LIGHTCOND':'Dusk'}, inplace=True)
dummy_variable_3.rename(columns={'LIGHTCOND':'Dawn', 'LIGHTCOND':'Dark - No Street Lights'}, inplace=True)
dummy_variable_3.rename(columns={'LIGHTCOND':'Dark - Street Lights Off', 'LIGHTCOND':'Other'}, inplace=True)
dummy_variable_3.rename(columns={'LIGHTCOND':'Dark - Unknown Lighting'}, inplace=True)

dummy_variable_4.rename(columns={'SPEEDING':'Y'}, inplace=True)

dummy_variable_1.head()
dummy_variable_2.head()
dummy_variable_3.head()
dummy_variable_4.head()

```

```python
# We now have i.e. the value 0 to represent "Dry" and 1 to represent "Wet" in the column "WEATHER" etc. 
# We will now insert this column back into our original dataset.

# merge data frame "cdf" and "dummy_variable_1" 
cdf = pd.concat([cdf, dummy_variable_1], axis=1)

# drop original column "WEATHER" from "cdf" 
cdf.drop("WEATHER", axis = 1, inplace=True)

# merge data frame "cdf" and "dummy_variable_1" 
cdf = pd.concat([cdf, dummy_variable_1], axis=1)


# drop original column "ROADCOND" from "cdf" 
cdf.drop("ROADCOND", axis = 1, inplace=True)

# merge data frame "cdf" and "dummy_variable_1" 
cdf = pd.concat([cdf, dummy_variable_1], axis=1)

# drop original column "LIGHTCOND" from "cdf" 
cdf.drop("LIGHTCOND", axis = 1, inplace=True)

# merge data frame "cdf" and "dummy_variable_1" 
cdf = pd.concat([cdf, dummy_variable_1], axis=1)

# drop original column "SPEEDING" from "cdf" 
cdf.drop("SPEEDING", axis = 1, inplace=True)



cdf.head()
```
